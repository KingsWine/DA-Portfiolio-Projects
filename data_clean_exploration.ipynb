# Data Cleaning and exploration Project

### Step 1: importing modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2: Importing data and viewing top rows

```python
auto_mo = pd.read_csv('automobile_data.csv')
auto_mo.head()
```

### Step 3: Data cleaning; checking for missing values and Duplicate

```python
auto_mo.isna()
```

```python
auto_mo.isna().any()
```

```python
auto_mo.isna().sum()
```

```python
print(auto_mo['num_of_doors'])
```

### Forward filling the null values
##### We found out that the null values can be filled with previous values

```python
auto_mo = auto_mo.fillna(method='ffill')
```

```python
auto_mo.isna().sum()
```

```python
auto_mo.duplicated().sum()
```

```python
auto_mo = auto_mo.drop_duplicates()
auto_mo.info()
```

```python
auto_mo.describe()
```

```python
auto_mo.columns
```

```python
auto_mo.head()
```

```python
print(auto_mo.dtypes)
```

```python
auto_mo['price'] = auto_mo['price'].str.replace('\.00', '')
```

```python
auto_mo['price'] = auto_mo['price'].str.replace('$', '')
print(auto_mo['price'])
```

```python
auto_mo['price'] = auto_mo['price'].str.replace(',', '')
```

```python
print(auto_mo['price'])
```

```python
auto_mo['price'] = auto_mo['price'].astype(int)
```

```python
auto_mo.describe()
```

```python
auto_mo['price'].sum()
```

### Step 4: Finding the number of car makes and their total price

```python
car_make_count = auto_mo['make'].value_counts(sort=True)
print(car_make_count)
```

```python
car_make_price = auto_mo.groupby('make')['price'].sum()
print(car_make_price)
```

```python
car_make_price.to_csv('car_make.csv')
```

### Step 5: Filter the body_style column, create a label column and find the frequency between interest and others

```python
auto_mo['label'] = np.where(auto_mo['body_style'] == 'convertible', 'interest', 'others')
```

```python
print(auto_mo['label'])
```

```python
label_count = auto_mo['label'].value_counts()
print(label_count)
```

### Step 6: Finding the most and least expensive car by make and ploting a graph

```python
car_make = auto_mo.groupby('make')['price'].sum()
print(car_make)
```

```python
car_make.plot(kind = 'bar', title = 'Car Price by Make')
plt.show()
```
